# 전화번호 프로그램 만들기

* 홍길동 010-123-4567
* 기능1 : 연락처 추가
* 기능2 : 연락처 전체 보기
* 기능3 : 검색, 이름을 입력받아서 전화번호 조회
* 기능4 : 수정, 이름을 입력받아서 전화번호 입력수정
* 기능5 : 삭제, 이름 입력받아서 삭제
* 기능6 : 프로그램 종료

```python
contacts = {}  # 전화번호부를 저장할 딕셔너리

while True:
    print("===== 전화번호부 프로그램 =====")
    print("1. 연락처 추가")
    print("2. 연락처 전체 보기")
    print("3. 검색")
    print("4. 수정")
    print("5. 삭제")
    print("6. 프로그램 종료")
    
    선택 = input("원하는 기능의 번호를 입력: ")
    
    if 선택 == '1':
        name = input("이름 입력: ")
        number = input("전화번호 입력: ")
        contacts[name] = number
        print("연락처가 추가됨.")
    
    elif 선택 == '2':
        print("===== 연락처 전체 보기 =====")
        for name, number in contacts.items():
            print(f"{name}: {number}")
    
    elif 선택 == '3':
        name = input("검색할 이름을 입력: ")
        number = contacts.get(name)
        if number:
            print(f"{name}의 전화번호: {number}")
        else:
            print("해당하는 연락처가 없음.")
    
    elif 선택 == '4':
        name = input("수정할 이름을 입력: ")
        number = contacts.get(name)
        if number:
            new_number = input("새로운 전화번호를 입력: ")
            연락처[name] = new_number
            print("전화번호가 수정됨.")
        else:
            print("해당하는 연락처가 없음.")
    
    elif 선택 == '5':
        name = input("삭제할 이름을 입력: ")
        number = contacts.get(name)
        if number:
            del contacts[name]
            print("연락처 삭제됨.")
        else:
            print("해당 연락처가 없음.")
    
    elif 선택 == '6':
        print("프로그램 종료.")
        break
    
    else:
        print("잘못된 입력. 다시 입력하시오.")
```
